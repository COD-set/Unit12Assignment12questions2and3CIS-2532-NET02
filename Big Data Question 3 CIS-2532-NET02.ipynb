```python
import sqlite3
connection = sqlite3.connect('books.db')
import pandas as pd
pd.options.display.max_columns = 10
```

```python
pd.read_sql("""SELECT title, edition, copyright
                    FROM titles""", connection)
```

```python
#read authors last names in descending order
pd.read_sql("""SELECT id,  last
                    FROM authors
                    ORDER BY id DESC""",
                 connection, index_col=['id'])

```

```python
#all book titles in ascending order
pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
                  connection)
```

```python
#lets see all the columns we are working with for the author 
pd.read_sql('SELECT * FROM authors', connection)
```

```python
#all the columns for author_ISBN
pd.read_sql('SELECT * FROM author_ISBN', connection)
```

```python
#all the columns for titles
pd.read_sql('SELECT * FROM titles', connection)
```

```python
#now we will merge data using the INNER JOIN and display books written by Paul Deitel who has an author id of 1

pd.read_sql("""SELECT titles.isbn, title, copyright
                    FROM titles
                    INNER JOIN author_ISBN
                        ON titles.isbn = author_ISBN.isbn
                        WHERE author_ISBN.id = '1'
                        ORDER BY title""", connection)
```

```python
#now we will initialize the cursor and insert a new author! Their name is Ajax Bellweather
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last)
                                VALUES ('Ajax', 'Bellweather')""")
```
